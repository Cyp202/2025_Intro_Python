````python
# Perform silhouette analysis following instructions in the tutorial notebook
silhouette_avg = [] #a function
for num_clusters in range(2,10): #On ne teste pas k=1, car la silhouette n’a pas de sens avec un seul cluster.
  # initialise kmeans
  kmeans = KMeans(n_clusters=num_clusters) #Tu initialises un modèle K-Means avec le nombre de clusters courant (num_clusters).
  kmeans.fit(X) #Tu entraînes le modèle sur les données X → K-Means trouve les centres des clusters et affecte chaque point à l’un d’eux.
  cluster_labels = kmeans.labels_ #paramètre, #cela va nous dire dans quel cluster nos nouvelles data va aller
  # silhouette score
  silhouette_avg.append(silhouette_score(X, cluster_labels)) #calcule le score de silhouette, une mesure de la qualité du clustering.

#______________ = []
#for num_clusters in range(2,10):
  # initialise kmeans
 # ______ = _______(___________)
 # ______.fit(X)
 # cluster_labels = _______.__________

  # silhouette score
  #_____________.append(silhouette_score(_, ________________))
````

````python
# Plot your silhouette analysis result here.
plt.plot(range(2,10),silhouette_avg,marker='s',c='k',lw=2)
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis for Optimal K')
plt.show()
````

### Q5: Train the KMeans clustering with 2 clusters

````python
# Train k-means clustering algorithm here
kmeans = KMeans(n_clusters=2)
kmeans.fit(X) #la position des centres des clusters (les "centroids"),la logique de séparation des points entre ces clusters.

# Predict labels
y_kmeans = kmeans.predict(X) #va attribuer à chaque point de ton jeu de données X le numéro du cluster auquel il appartient.
penguin_df['cluster_predicted2'] = y_kmeans # Cette ligne ajoute les étiquettes de cluster (y_kmeans) dans ton dataframe penguin_df sous une nouvelle colonne appelée "label2".
````
